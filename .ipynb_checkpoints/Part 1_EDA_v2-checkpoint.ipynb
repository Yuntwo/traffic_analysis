# 1. Introduction

For this project there are three questions which will be analyzed.
1. Is there any relationship between the number of electric vehicle adopted with changes in air quality
2. Is there any relationship between the various types of fuel for land vehicles with changes in air quality
3. Is there any relationship between traffic condition/flow with air quality

Every part of the world is finding ways to mitigate climate change and Singapore is not an exception. In October 2022, Singapore announced they will
aim for net zero emission by the year 2050. This target is accompanied by several efforts which one of them is the adoption of electric vehicles (EV).
Singapore plans to add incentives by adopting EV and further improve EV infrastructure in the country.

All of this effort to push EV adoption raises the question whether EV makes a difference in helping climate change. One metric that can help answer the question is air quality. Pollutants caused by vehicles emissions contribute to the climate change. The idea to reduce this pollutant is where EV comes in. EV does not create tailpipe commissions but it does create emissions from the power grid. We want to see if there is any relationship between adopting EV with air quality since non-existent tail pipe emission maybe offset by the emissions created by power grid. We also want to see the possibility of other fuel type contribution to the air quality.

To strengthen the conclusion from this hypotheses, we pick another city of reference. For this project we pick London as the city to compare Singapore with. The reason is London has similar plans regarding EV for the future. London plans to ramp its EV infrastructure by 2030 which is the same timeframe with Singapore.

```python
sg_pivot_df['date'] = pd.to_datetime(sg_pivot_df['date'])

# Aggregate data to monthly averages
monthly_df = sg_pivot_df.set_index('date').resample('M').mean().reset_index()

# Define the function to plot air quality metrics
def plot_air_quality_metrics(df, air_quality_columns):
    # Extract regions from column names
    regions = set()
    for col in air_quality_columns:
        parts = col.split('_')
        if len(parts) >= 3:  # Check if there are enough parts
            regions.add(parts[2])  # Add the region part

    # Create a separate plot for each region
    for region in regions:
        # Filter columns for the current region
        region_columns = [col for col in air_quality_columns if f'_{region}' in col]
        
        if region_columns:
            plt.figure(figsize=(15, 8))
            for col in region_columns:
                sns.lineplot(data=monthly_df, x='date', y=col, label=col)

            plt.title(f'Air Quality Metrics for {region.capitalize()} Region (Monthly Averages)')
            plt.xlabel('Date')
            plt.ylabel('Values')
            plt.legend()
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()

# List of air quality columns
air_quality_columns = [
    'co_central', 'co_east', 'co_national', 'co_north', 'co_south',
       'co_west', 'o3_central', 'o3_east', 'o3_national', 'o3_north',
       'o3_south', 'o3_west', 'pm10_central', 'pm10_east', 'pm10_national',
       'pm10_north', 'pm10_south', 'pm10_west', 'pm25_central', 'pm25_east',
       'pm25_national', 'pm25_north', 'pm25_south', 'pm25_west',
       'psi_24hr_central', 'psi_24hr_east', 'psi_24hr_national',
       'psi_24hr_north', 'psi_24hr_south', 'psi_24hr_west', 'so2_central',
       'so2_east', 'so2_national', 'so2_north', 'so2_south', 'so2_west'
]

plot_air_quality_metrics(monthly_df, air_quality_columns)
```

Based on this grouping, we observe again the trend of the number of vehicles. The trend is similar to what we have plotted before before the grouping. ## 4.1 Air Quality Improvement Hypothesis:

***Null Hypothesis (H0):*** There is no significant relationship between the increase in EV adoption and changes in air quality.

***Alternative Hypothesis (H1):*** There is a significant relationship between the increase in EV adoption and improvements in air quality (measured by reductions in (CO, SO2, O3, PM2.5, PM10 and PSI values)).


```python
# Step 1: Convert date columns to datetime format and extract year-month for merging
air_quality_data['year_month'] = pd.to_datetime(air_quality_data['date']).dt.to_period('M')
vehicle_population['year_month'] = pd.to_datetime(vehicle_population['month']).dt.to_period('M')

# Step 2: Filter the vehicle population for EV and non-EV cars only
ev_population = vehicle_population[vehicle_population['type'] == 'Electric']
non_ev_population = vehicle_population[vehicle_population['type'] != 'Electric']

# Step 3: Aggregate EV and non-EV vehicle population by year-month
ev_grouped = ev_population.groupby('year_month')['number'].sum().reset_index(name='EV')
non_ev_grouped = non_ev_population.groupby('year_month')['number'].sum().reset_index(name='Non_EV')

# Step 4: Merge the air quality data with EV and Non-EV population
merged_data = pd.merge(air_quality_data, ev_grouped, on='year_month', how='left')
merged_data = pd.merge(merged_data, non_ev_grouped, on='year_month', how='left')

# Preview the merged data
merged_data.head()
```

```python
scaler = MinMaxScaler()

# Normalize the air quality metrics (CO, PM2.5, PM10, PSI)
cols_to_normalize = ['EV']
merged_data[cols_to_normalize] = scaler.fit_transform(merged_data[cols_to_normalize])

# Preview the normalized data
merged_data.head()
```

```python
# Calculate correlation between EV adoption and air quality metrics
correlation_matrix = merged_data[['EV', 'co_national', 'pm25_national', 'pm10_national','o3_national','so2_national' ,'psi_24hr_national']].corr()
sns.heatmap(correlation_matrix,annot=True)
plt.show()
```

### Hypothesis Test Method 1:
***Spearman Correlation***: This test is used to assess the linear relationship between the increase in EV adoption and changes in air quality metrics. Since we want to explore whether there is a monotonic relationship between the variables, Spearman's correlation is appropriate here. We test the significance of this relationship by examining the p-values.

```python
from scipy.stats import spearmanr

# Perform Spearman correlation test for each air quality metric against EV numbers
co_spearman_corr, co_spearman_pval = spearmanr(merged_data['EV'], merged_data['co_national'])
o3_spearman_corr, o3_spearman_pval = spearmanr(merged_data['EV'], merged_data['o3_national'])
so2_spearman_corr, so2_spearman_pval = spearmanr(merged_data['EV'], merged_data['so2_national'])
pm25_spearman_corr, pm25_spearman_pval = spearmanr(merged_data['EV'], merged_data['pm25_national'])
pm10_spearman_corr, pm10_spearman_pval = spearmanr(merged_data['EV'], merged_data['pm10_national'])
psi_spearman_corr, psi_spearman_pval = spearmanr(merged_data['EV'], merged_data['psi_24hr_national'])

# Print Spearman correlation results
print(f"CO vs EV: Spearman Corr = {co_spearman_corr}, p-value = {co_spearman_pval}")
print(f"O3 vs EV: Spearman Corr = {o3_spearman_corr}, p-value = {o3_spearman_pval}")
print(f"So2 vs EV: Spearman Corr = {so2_spearman_corr}, p-value = {so2_spearman_pval}")
print(f"PM2.5 vs EV: Spearman Corr = {pm25_spearman_corr}, p-value = {pm25_spearman_pval}")
print(f"PM10 vs EV: Spearman Corr = {pm10_spearman_corr}, p-value = {pm10_spearman_pval}")
print(f"PSI vs EV: Spearman Corr = {psi_spearman_corr}, p-value = {psi_spearman_pval}")
```

```python
def hypothesis_test(corr, p_val, alpha=0.05):
    if p_val < alpha:
        return f"Reject the null hypothesis. Significant relationship detected (Correlation: {corr}, p-value: {p_val})."
    else:
        return f"Fail to reject the null hypothesis. No significant relationship detected (Correlation: {corr}, p-value: {p_val})."

# Hypothesis test results
co_result = hypothesis_test(co_corr, co_pval)
o3_result = hypothesis_test(o3_corr, o3_pval)
so2_result = hypothesis_test(so2_corr, so2_pval)
pm25_result = hypothesis_test(pm25_corr, pm25_pval)
pm10_result = hypothesis_test(pm10_corr, pm10_pval)
psi_result = hypothesis_test(psi_corr, psi_pval)

print(f"CO vs EV: {co_result}")
print(f"O3 vs EV: {o3_result}")
print(f"So2 vs EV: {so2_result}")
print(f"PM2.5 vs EV: {pm25_result}")
print(f"PM10 vs EV: {pm10_result}")
print(f"PSI vs EV: {psi_result}")
```

## 4.2 Fuel Type and Air Quality Hypothesis:

***Null Hypothesis (H0):*** The type of fuel used (petrol, diesel, electric) in vehicles has no significant impact on air quality.

***Alternative Hypothesis (H1):*** The type of fuel used in vehicles significantly impacts air quality.

```python
def categorize_fuel_type(fuel_type):
    if fuel_type in ['Petrol', 'Petrol-Electric', 'Petrol-Electric (Plug-In)', 'Petrol-CNG']:
        return 'Petrol-based'
    elif fuel_type in ['Diesel', 'Diesel-Electric', 'Diesel-Electric (Plug-In)']:
        return 'Diesel-based'
    elif fuel_type == 'CNG':
        return 'CNG'
    elif fuel_type == 'Electric':
        return 'Electric'

# Apply the function to categorize fuel types
vehicle_df['fuel_category'] = vehicle_df['type'].apply(categorize_fuel_type)
vehicle_df.head()
```

```python
# Merge vehicle data with air quality data
combined_df = pd.merge(air_quality_data[['year_month','co_national', 'o3_national','pm25_national', 'pm10_national', 'so2_national', 'psi_24hr_national']], vehicle_df, on='year_month')
combined_df.head()
```

```python
# Calculate weighted averages for each fuel category based on the number of vehicles
def weighted_mean(group, metric):
    return (group[metric] * group['number']).sum() / group['number'].sum()

# Group by fuel category and calculate weighted means for air quality metrics
weighted_grouped_df = combined_df.groupby('fuel_category').apply(
    lambda group: pd.Series({
        'co_national': weighted_mean(group, 'co_national'),
        'o3_national': weighted_mean(group, 'o3_national'),
        'pm25_national': weighted_mean(group, 'pm25_national'),
        'pm10_national': weighted_mean(group, 'pm10_national'),
        'so2_national': weighted_mean(group, 'so2_national'),
        'psi_24hr_national': weighted_mean(group, 'psi_24hr_national')
    })
).reset_index()
```

#### Weighted ANOVA Testing Methodology:
Accounting for Different Group Sizes (Vehicle Numbers):
* In the context of analyzing the impact of fuel type on air quality, the number of vehicles running on each fuel type varies significantly (e.g., far more petrol vehicles than electric vehicles). A standard ANOVA treats each observation equally, but this would ignore the fact that some fuel types have far more vehicles contributing to pollution.
* Weighted ANOVA addresses this by adjusting the analysis to reflect the true contribution of each fuel type to air quality based on how many vehicles of that type are on the road. Thus, categories with more vehicles are given more weight in the analysis, aligning the results with reality.

```python
# Prepare the data for ANOVA
fuel_types = ['Petrol-based', 'Diesel-based', 'Electric', 'CNG']
air_quality_metrics = ['co_national', 'o3_national','pm25_national', 'pm10_national', 'so2_national', 'psi_24hr_national']

# Perform ANOVA for each air quality metric
for metric in air_quality_metrics:
    data = [combined_df[combined_df['fuel_category'] == fuel][metric] for fuel in fuel_types]
    f_stat, p_value = stats.f_oneway(*data)
    print(f'ANOVA for {metric}: F-statistic = {f_stat}, p-value = {p_value}')

    # Hypothesis test interpretation
    if p_value < 0.05:
        print(f'Reject the null hypothesis for {metric}. There is a significant impact of fuel type on {metric}.')
    else:
        print(f'Fail to reject the null hypothesis for {metric}. There is no significant impact of fuel type on {metric}.')
```

### The results of the ANOVA tests indicate the following:

1. CO Levels (co_national): There is not enough evidence to reject the null hypothesis. This suggests that there is no significant impact of vehicle fuel type on CO levels. In other words, variations in fuel type (Petrol, Diesel, Electric, etc.) do not appear to influence CO concentrations in the air significantly.

2. Ozone Levels (o3_national): There is not enough evidence to reject the null hypothesis. This indicates that the fuel type used in vehicles does not have a significant impact on ozone (O3) levels.

3. PM2.5 Levels (pm25_national): There is not enough evidence to reject the null hypothesis. This suggests that vehicle fuel type does not significantly impact PM2.5 concentrations.

4. PM10 Levels (pm10_national): There is not enough evidence to reject the null hypothesis. The fuel type does not seem to significantly affect PM10 levels.

5. SO2 Levels (so2_national): ***We reject the null hypothesis.*** This indicates that there is a significant impact of vehicle fuel type on SO2 levels. The higher F-statistic and very low p-value suggest that differences in fuel type, particularly Diesel, have a notable influence on sulfur dioxide concentrations, possibly due to diesel engines emitting more sulfur.

6. PSI (psi_24hr_national): ***We reject the null hypothesis.*** This suggests that fuel type has a significant impact on the PSI (Pollutant Standards Index), which is a measure of air quality. The F-statistic indicates a notable variance in air quality based on fuel type.


* SO2 and PSI Levels: The significant p-values for SO2 and PSI indicate that certain fuel types, especially those with high sulfur content (like diesel), are more likely to impact these pollutants. This is consistent with what we know about diesel engines, which tend to emit more sulfur dioxide compared to petrol or electric vehicles.

* No Impact on Other Pollutants: For CO, ozone, PM2.5, and PM10 levels, there is no statistically significant impact from the type of fuel used. This may suggest that other factors, such as overall vehicle use, traffic congestion, or industrial emissions, might play a larger role in determining these pollutant levels.

***To further investigate the differences between fuel types for the air quality metrics that showed significant impact (i.e., SO2 and PSI), we will perform a post-hoc Tukey's HSD (Honestly Significant Difference) test. This test will help determine which specific fuel types differ significantly from each other.***

```python
import pandas as pd
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# and we are particularly interested in 'so2_national' and 'psi_24hr_national' which showed significant results.

# Weight the 'so2_national' and 'psi_24hr_national' by the number of vehicles in each fuel category
combined_df['so2_weighted'] = combined_df['so2_national'] * combined_df['number']
combined_df['psi_weighted'] = combined_df['psi_24hr_national'] * combined_df['number']

# Filter the data for SO2 levels (weighted)
so2_weighted_data = combined_df[['fuel_category', 'so2_weighted']].dropna()

# Perform Tukey HSD post-hoc test for SO2 levels (weighted)
tukey_so2_weighted = pairwise_tukeyhsd(endog=so2_weighted_data['so2_weighted'], 
                                       groups=so2_weighted_data['fuel_category'], 
                                       alpha=0.05)

# Display the results for SO2 levels (weighted)
print('Tukey HSD Test for Weighted SO2 Levels:')
print(tukey_so2_weighted)

# Filter the data for PSI levels (weighted)
psi_weighted_data = combined_df[['fuel_category', 'psi_weighted']].dropna()

# Perform Tukey HSD post-hoc test for PSI levels (weighted)
tukey_psi_weighted = pairwise_tukeyhsd(endog=psi_weighted_data['psi_weighted'], 
                                       groups=psi_weighted_data['fuel_category'], 
                                       alpha=0.05)

# Display the results for PSI levels (weighted)
print('\nTukey HSD Test for Weighted PSI Levels:')
print(tukey_psi_weighted)
```
